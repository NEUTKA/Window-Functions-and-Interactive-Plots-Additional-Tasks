# Task 1  
## "What is the most frequent value of the `Parent` attribute?"

```python
# Reading the Data File
df = pd.read_csv('D:/курс_архив/my_files/lesson_6/gff.tsv', sep="\t")
df
```

```python
# Extracting the `attributes` Column
attributes_column = df.attributes
```

```python
# Converting to DataFrame for Inspection
pd.DataFrame(attributes_column)
```

```python
# Regular Expression for Extracting Values After "Parent="
# \w+ — Matches a sequence of alphanumeric characters (identifier)
find_words_after_parent = re.compile(r'Parent=(\w+)')
find_words_after_parent 
```

```python
# Applying the Regular Expression to All Rows in the 'attributes' Column Using `apply()`
# `findall()` returns a list of all matches; we take the first match [0] if it exists, otherwise None
df['Parent'] = df['attributes'].apply(lambda x: find_words_after_parent.findall(x)[0] if find_words_after_parent.findall(x) else None)
print(df[['attributes', 'Parent']]) # Displaying the DataFrame with Two Columns: 'attributes' and 'Parent' for Verification
```

```python
# Using `value_counts()` to Count the Frequency of Each Value in the 'Parent' Column
# `idxmax()` Returns the Most Frequent Value
most_frequent_parent = df['Parent'].value_counts().idxmax()
# Output the Most Frequent Value
print(f"The Most Frequent Value of the 'Parent' Attribute: {most_frequent_parent}")
```
